```python
occupancy = pd.read_csv('../data/us_occupancy_data.csv')
occupancy['Occupancy_2'] = occupancy['Occupancy Rate'].str[:-1]
occupancy
```

```python
occupancy['Occupancy_2'] = occupancy['Occupancy_2'].astype(int)
```

```python
occupancy.info()
```

```python
occupancy['Occupancy Rate'].sort_values(ascending=False).head(15)
```

```python
list = [93,87,80,79,77,75,74,73,72]
top20 = occupancy.loc[occupancy['Occupancy_2'].isin(list)]
top20 = top20.sort_values(by='Occupancy_2', ascending = False)
```

```python
top20
```

```python
plt.barh(y=top20['State '], width=top20['Occupancy_2'], color = 'red')
plt.title('Highest Occupancy by State')
plt.xlabel('Occupancy Rate(percentage)')
```

```python
occupancy['Annual Revenue_2'] = occupancy['Annual Revenue'].str[1:]
occupancy['Annual Revenue_2'] = occupancy['Annual Revenue_2'].str.replace(',' , '')
top15_revenue = occupancy[['Annual Revenue_2', 'State ']].sort_values(by = 'Annual Revenue_2', ascending = False)
```

```python
top15_revenue['Annual Revenue_2'] =  top15_revenue['Annual Revenue_2'].astype(int)
```

```python
top15_revenue = top15_revenue.reset_index(drop=True)
top15_revenue = top15_revenue.sort_values(by='Annual Revenue_2', ascending = False).reset_index(drop=True).head(20)
top15_revenue
```

```python
plt.barh(y=top15_revenue['State '], width=top15_revenue['Annual Revenue_2'], color = 'red')
plt.title('Top Revenues By State')
plt.xlabel('Revenue')
```
